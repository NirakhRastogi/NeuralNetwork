```python
import numpy as np
import matplotlib.pyplot as plt
```

Mean Squared Error, l = (dy - py)^2
---
(Desired) dy = 2*x+1

(Predicted) py= 3*x-1



```python
x = [t for t in range(-50,50)]
dy = [2*t+1 for t in x]
py= [3*t-1 for t in x]
```

```python
l = [(d-p)**2 for d,p in zip(dy,py)]
```

```python
plt.figure()
plt.plot(x, l)
plt.show()
```

```python
plt.figure()
plt.scatter(x,dy)
plt.scatter(x,py)
plt.show()
```
