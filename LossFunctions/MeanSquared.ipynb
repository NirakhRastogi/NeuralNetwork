```python
import numpy as np
import matplotlib.pyplot as plt
```

Mean Squared Error, l = (dy - py)^2
---
(Desired) dy = 2*x+1

(Predicted) py= 3*x-1



```python
x = [t for t in range(-50,50)]
dy = [2*t+1 for t in x]
py= [3*t-1 for t in x]
```

```python
l = [(d-p)**2 for d,p in zip(dy,py)]
```

```python
plt.figure()
plt.title('Loss Plot')
plt.plot(x, l)
plt.show()
```

```python
plt.figure()
plt.title('Fig. 2')
plt.plot(x,dy)
plt.plot(x,py)
for (_x,_l) in zip(x,l):
    plt.plot([_x,_x],[0,(_l*max(max(dy), max(py)))/max(l)], c='gray')
plt.show()
```

Fig. 2, shows the loss to be more at the extreme ends where the differencce between the original(dy) and predicted(py) value is more. And, by square factor of the loss, even the difference between dy-py is -ve, we see the loss to be +ve. This will make our model to consider the loss if there is a deviation in the desired and predicted values irrespective of the direction of deviation.
